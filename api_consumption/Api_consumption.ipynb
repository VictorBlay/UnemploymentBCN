```python
import requests
```

```python
url = "https://opendata-ajuntament.barcelona.cat/data/api/3/action/package_search"
query = {"rows":"3000"}
```

```python
res = requests.get(url, params = query)
res
```

```python
url_tornada = res.json()["result"]["results"]
```

```python
lista_url = [r["url_tornada"]["es"] for r in url_tornada]
```
